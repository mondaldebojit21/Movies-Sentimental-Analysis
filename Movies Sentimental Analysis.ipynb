```python
#fetching data by directly scraping the content from the web pages
#importing libraries for the same
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# creating pandas data frame for storing the data
df=pd.DataFrame()
name=[]
link=[]
review_comment=[]
review_rating_percentage=[]
```

```python
# Accessing the HTML content from webpage to fetch movies link
source = requests.get('https://www.rottentomatoes.com/browse/movies_at_home/?page=1000')

# Parsing the HTML content 
soup = BeautifulSoup(source.text, 'html.parser')

# Searching and navigating through content to fetch the movie links 
movies = soup.find('div', class_='discovery-grids-container')
k = movies.find("div", class_='discovery-tiles').find_all('a')

# Running loop in order to fetch data for the number of movie links
for j in k:

    # Accessing the respective movie reviews HTML content
    ref = 'https://www.rottentomatoes.com' + j.get('href') + '/reviews?type=user'
    source = requests.get(ref)
    soup = BeautifulSoup(source.text, 'html.parser')
    

    
    # Checking if reviews are present or not, if not then printing the link
    try:
        # Searching and navigating through the parse content to fetch the reviews and rating 
        comments = soup.find('ul', class_='audience-reviews').find_all("li", {"class": "audience-reviews__item"})
        
        
        # Searching and Storing all the audience reviews and the rating for the respective movie
        for i in comments:
            
            # Appending the movie name in the list
            name.append(j.text.strip())
            # Appending the respective movie reviews HTML link
            link.append(ref)
            
            # geting the 
            review_comment.append(i.find('p', class_='audience-reviews__review js-review-text clamp clamp-8 js-clamp').text)
            
            
            # Calculating the review percentage based on 5 star
            fill_star_count = len(i.find_all('span', {"class": "star-display__filled"}))
            half_star_count = len(i.find_all('span', {"class": "star-display__half"}))
            review_rating_percentage.append((fill_star_count + half_star_count) * 100 / 5)
                       
    except Exception as e:
        print(e,ref)
#         print('No Reviews for ',j.text.strip(),' refer link ',ref)

        
print('Completed')
df['Movies']=name
df['Links']=link
df['Review Comment']=review_comment
df['Review Rating %']=review_rating_percentage
```

```python
mydata=df.copy()
```

```python
import re

def clean(text):
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text

mydata['Cleaned Reviews'] = mydata['Review Comment'].apply(clean)
mydata.head()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.corpus import brown
```

```python
mydata['Tokenized'] = mydata['Cleaned Reviews'].apply(word_tokenize)
```

```python
mydata
```

```python
def remove_stopwords(text):
    st=[]
    for i in text:
        if i.lower() not in stopwords.words('english'):
            st.append(i)
    return st
mydata['Tokenized'] = mydata['Tokenized'].apply(remove_stopwords)
mydata['POS_Tag']=mydata['Tokenized'].apply(pos_tag)
mydata
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}

def lemmatize(pos_data):
    lemma = []
    for word, pos in pos_data:
        if not pos_dict.get(pos[0]):
            lemma.append(word)
        else:
            lemma.append(wordnet_lemmatizer.lemmatize(word, pos_dict.get(pos[0])))
    return (' ').join(lemma)

mydata['Lemma'] = mydata['POS_Tag'].apply(lemmatize)
mydata
```

```python
from textblob import TextBlob
```

```python
# function to calculate subjectivity
def getSubjectivity(review):
    return TextBlob(review).sentiment.subjectivity

# function to calculate polarity
def getPolarity(review):
    return TextBlob(review).sentiment.polarity

# function to analyze the reviews
def analysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
# function to calculate polarity %
def getPolarity_percentage(review):
    return ((TextBlob(review).sentiment.polarity)+1)*100/2
```

```python
# fin_data['Subjectivity'] = fin_data['Lemma'].apply(getSubjectivity) 
mydata['Polarity'] = mydata['Lemma'].apply(getPolarity) 
mydata['Analysis'] = mydata['Polarity'].apply(analysis)
mydata['Polarity %'] = mydata['Lemma'].apply(getPolarity_percentage)
```
